````python

import numpy as np
import pymc as pm
import arviz as az

# Generate data
np.random.seed(42)
true_theta = 5
true_sigma = 2
data = np.random.normal(true_theta, true_sigma, 100)

with pm.Model() as model1:
    # Priors
    theta = pm.Normal('theta', mu=0, sigma=10)
    sigma = pm.Gamma('sigma', alpha=2, beta=1)
    # Likelihood
    y_obs = pm.Normal('y_obs', mu=theta, sigma=sigma, observed=data)
    # Sampling
    trace1 = pm.sample(2000, tune=2000)

# Diagnostics
az.summary(trace1)
az.plot_trace(trace1)
az.plot_autocorr(trace1)
````

````python
with pm.Model() as model2:
    theta = pm.Laplace('theta', mu=0, b=1)
    sigma = pm.LogNormal('sigma', mu=0, sigma=1)
    y_obs = pm.Normal('y_obs', mu=theta, sigma=sigma, observed=data)
    trace2 = pm.sample(2000, tune=2000)

# Diagnostics
az.summary(trace2)
az.plot_trace(trace2)
az.plot_autocorr(trace2)
````

````python
with pm.Model() as model3:
    theta = pm.Cauchy('theta', alpha=0, beta=1)
    sigma = pm.Exponential('sigma', lam=0.5)  # lam=0.5 gives mean=2
    y_obs = pm.Normal('y_obs', mu=theta, sigma=sigma, observed=data)
    trace3 = pm.sample(2000, tune=2000, target_accept=0.9)

# Diagnostics
az.summary(trace3)
az.plot_trace(trace3)
az.plot_autocorr(trace3)
````

# Slice Sampling & Metropolis-within-Gibbs

---


### **What It Does**  
Samples values from a probability distribution (e.g., a bell curve) **without needing to know its total area**.  

### **How It Works**  
Imagine drawing a horizontal line under the curve. Then:  
1. **Draw a random height** for the line (slice level `y`) between 0 and the current points height.  
2. **Find the horizontal region** where the curve is above this line.  
3. **Pick a random point** in this region. If its valid (still above the line), keep it. If not, shrink the region and try again.  

---

## **2. Using Slice Sampling in Gibbs Sampling**

### **The Problem**  
In Gibbs sampling, we update parameters one by one. If a parameters update rule is messy (e.g., no formula), Metropolis-Hastings (MH) is used but requires tuning.  

### **Slice Sampling Fixes This**  
- **No Tuning Needed:** Just use the unnormalized probability (no exact formula required).  
- **Steps for Each Parameter:**  
  1. **Draw a slice level `y`** under the current parameters probability.  
  2. **Find the interval** where the probability  `y`.  
  3. **Sample a new value** from this interval. No rejections!
